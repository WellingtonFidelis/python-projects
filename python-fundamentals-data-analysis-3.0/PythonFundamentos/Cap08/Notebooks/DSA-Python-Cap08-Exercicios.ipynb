### Exercício 6

```python
# Considere os 3 arrays abaixo
# Retorne o valor do array xarr se o valor for True no array cond. Caso contrário, retorne o valor do array yarr.
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
result = [(arrayx if conditional else array) for arrayx, array, conditional in zip(xarr, yarr, cond)]
result
```

```python
array_temp2 = zip(xarr, yarr, cond)
for item in array_temp2:
    print(item)
print(array_temp2)
for (x, y, c) in zip(xarr, yarr, cond):
    if c:
        print(x)
    else:
        print(y)
```

### Exercício 10

```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros,
# consulte a tabela e retorne os dados em dataframe do Pandas
```

```python
import sqlite3

connection = sqlite3.connect('issues.db')
print('Opening database')
```

```python
cursor = connection.cursor()
```

```python
# Command DDL

sql_create_table = '''
CREATE TABLE IF NOT EXISTS issues (
    number INTEGER PRIMARY KEY NOT NULL,
    title TEXT NOT NULL,
    labels TEXT,
    state TEXT NOT NULL
);
'''
```

```python
# Executing instruction

cursor.execute(sql_create_table)
```

```python
# Iserting function
sql_insert = 'INSERT INTO issues VALUES (?, ?, ?, ?, ?);'
```

```python
# List with some records
record_set = [
    (0, 0, 'Breves respostas para grandes questões', 'Stephen Hawking', 'Not read'),
    (1, 1, 'O livro dos cinco anéis', 'Musashi', 'Reading'),
]
```

```python
#For each element of my list, save in database using the insert sql function
for record in record_set:
    cursor.execute(sql_insert, record)
```

```python
# Sending data to database
connection.commit()
```

```python
# Setting sql function to consult all records
sql_get_all = 'SELECT * FROM issues;'
```

```python
# Using the cursor to run all database and save the data in data variable
cursor.execute(sql_get_all)
data = cursor.fetchall()
```

```python
# For each record of the data, print the record
for item in data:
    print('Livro ID: (%d, %d), título: %s de autor: %s e status: %s.' % item)
```

```python
# Saving data into data frame of Pandas package
issues_df = pd.DataFrame(data, columns=['id', 'number', 'title', 'label', 'state'])

display(issues_df)
```
